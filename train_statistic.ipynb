```python
import re
i = 0
data = {}
import numpy as np

last_index = 0
with open('./screenlog.0') as f:
    for line in f:
        if line[:4] == 'INFO':
            regex = re.findall('(\d+(?:\.\d+)?)', line)
            if len(regex)!=2:
                continue
            index,time = regex
            t = line.split(' ')[1]
            if t not in data:
                data[t]=[]
            if index is -1:
                index = last_index
            else:
                last_index = index
            data[t].append(np.array([float(index), float(time)]))
        i+=1
        #if i>200:
         #   break
        
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
fig = plt.figure(figsize=(8, 12))
keys = data.keys()
for index, k in enumerate(keys):
    data[k] = np.array(data[k])
for index, k in enumerate(keys):
    ax = plt.subplot(len(keys),1,index+1)
    plt.tight_layout()
    ax.plot(data[k][:,0], data[k][:,1])
    ax.set_title(k)
```
