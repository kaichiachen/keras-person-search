```python
bbox = np.array([
         [533, 144, 600, 323]])
fig,ax = plt.subplots()
img=mpimg.imread('/home/workspace/keras-person-search/data/psdb/dataset/Image/SSM/s3473.jpg')
ax.imshow(img)
colors= ['r','g','b','y','w','m','k','c']
for i, box in enumerate(bbox):
    rect = patches.Rectangle((box[0],box[1]),box[2]-box[0],box[3]-box[1],linewidth=1,edgecolor=colors[i%len(colors)],facecolor='none')
    ax.add_patch(rect)

plt.show()
```
